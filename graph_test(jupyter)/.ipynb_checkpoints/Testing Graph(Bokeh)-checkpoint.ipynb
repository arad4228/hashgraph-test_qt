```python
import numpy as np
from bokeh.io import show
from bokeh.plotting import figure
from bokeh.models import Arrow, NormalHead, VeeHead

# 노드 개수
n_nodes = 5

# 랜덤한 방향성 그래프 생성
graph = np.random.randint(0, n_nodes, size=(n_nodes, n_nodes))
graph[graph != 0] = 1

# 노드 위치 지정
pos_x = [0, 1, 1, 1, 2]
pos_y = [0, 0, 1, 2, 1]

# 그래프 생성
p = figure(width=500, height=500, x_range=(-1, 3), y_range=(-1, 3))
p.circle(pos_x, pos_y, size=30, fill_color='white')

# 화살표 추가
for i in range(n_nodes):
    for j in range(n_nodes):
        if graph[i][j] == 1:
            arrow = Arrow(end=VeeHead(size=12), x_start=pos_x[i], y_start=pos_y[i], x_end=pos_x[j], y_end=pos_y[j])
            p.add_layout(arrow)

# 출력
show(p)
```
